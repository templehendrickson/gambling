```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()
```

Histogram and KDE of Normal vs Student T with 4 degrees of freedom (df)

```python
tArr = np.random.standard_t(4,10000)/100
nArr = np.random.normal(0,1,10000)/100

dist = pd.DataFrame({'studentT':tArr,'normal':nArr})

sns.histplot(dist,kde=True).set(title='norma vs student t distribution (4 df)')
```

### Student T Distribution

```python
tdf = pd.DataFrame({str(n) : np.random.standard_t(4,1000)/100 for n in range(10)})
```

```python
((tdf+1).cumprod()-1).plot(kind='line',title='10 sample from student t df=4',legend=False)
```

```python
(tdf.std()).mean()
```

Mean std of student t is higher, this makes sense as student t has fatter tails.

### Normal Distribution

```python
ndf = pd.DataFrame({str(n) : np.random.normal(0,1,1000)/100 for n in range(10)})
```

```python
((ndf+1).cumprod()-1).plot(kind='line',title='10 samples from normal, mu = 0, sigma = 1',legend=False)
```
